```python
import pandas as pd
import os
```

```python
### Get a list of MAF/VCF file names from the 50k UKB filtered by CH gene coordinates

# List files in directory
files = os.listdir('/workspace/datasets/ukbiobank_ch/vcf_ch_50k')

# Filter gvcf files
files = [file for file in files if '.maf.gz' in file]

#
print(files[:10])
print(len(files))
```

```python
### Create the job instructions for the 50k qmap jobs

# Set files and folders needed
python_script = '/workspace/projects/clonalhemato_ukb/analysis_50k_202103/analysis_ch_genes_vep_202104/filter_ch_variants.py'
input_folder = '/workspace/datasets/ukbiobank_ch/vcf_ch_50k/'
output_folder = '/workspace/projects/clonalhemato_ukb/ch_variants_50k/'

# Create list of commands
commands = []
for file in files:
    command = 'python ' + python_script + ' -i ' + input_folder + file + ' -o ' + output_folder
    commands.append(command)

# Check
print(len(commands))
commands
```

```python
### Set qmap file with parameters and jobs

qmap_pre_params = ['[pre]','. "/home/$USER/miniconda3/etc/profile.d/conda.sh"','conda activate clonalh','[params]','cores = 1','memory = 8G','[jobs]']
qmap_file = qmap_pre_params + commands
qmap_file
```
