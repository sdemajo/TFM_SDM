```python
import pandas as pd
import os
```

```python
### Get a list of MAF/VCF file names from the 50k UKB filtered by CH gene coordinates

# List files in directory
files = os.listdir('/workspace/projects/clonalhemato_ukb/ch_variants_50k/')

# Filter gvcf files
files = [file for file in files if '.maf.vep' in file]

#
print(files[:10])
print(len(files))
```

```python
### Create the job instructions for the 50k qmap jobs

# Set files and folders needed
python_script = '/workspace/projects/clonalhemato_ukb/analysis_50k_202103/analysis_ch_variants_202104/filterarrange_chvariants_mafvep.py'
input_folder = '/workspace/projects/clonalhemato_ukb/ch_variants_50k/'
output_folder = '/workspace/projects/clonalhemato_ukb/ch_variants_50k_mafvep/'

# Create list of commands
commands = []
for file in files:
    command = 'python ' + python_script + ' -i ' + input_folder + file + ' -o ' + output_folder
    commands.append(command)

# Check
print(len(commands))
commands
```

```python
### Set qmap file with parameters and jobs

qmap_pre_params = ['[pre]','. "/home/$USER/miniconda3/etc/profile.d/conda.sh"','conda activate clonalh','[params]','cores = 1','memory = 8G','[jobs]']
qmap_file = qmap_pre_params + commands
qmap_file
```

```python
### Save qmap file

with open('/workspace/projects/clonalhemato_ukb/ch_variants_50k_mafvep/filterarrange_chvariants.qmap', 'w') as f:
    for item in qmap_file:
        f.write('%s\n' % item)
```

```python
### TESTS 100

# qmap file test
qmap_file2 = qmap_pre_params + commands[:100]
qmap_file2


#Save qmap test file

with open('/workspace/projects/clonalhemato_ukb/ch_variants_50k_mafvep/test100/filterarrange_chvariants_test100.qmap', 'w') as f:
    for item in qmap_file2:
        f.write('%s\n' % item)
```
